### 3.1 

연령대(ex.10,20,30,40대) 변수와 함께 어떤 변수와의 관계를 보면 좋을지 기대 가설을 세우고, 적절한 방식으로 시각화 후 해석 보태 주세요.
연령은 25세, 26세~ 아닌 20대의 BIN 형태 연령대 (범주형 변수)로 묶어주세요.

[참고](https://nittaku.tistory.com/124)

```python
df=pd.read_csv("nyc_citibike.csv",encoding='euc-kr') # 한글이 들어간 csv는 encoding 인자를 넣어주는 것이 좋음
df.head()

#가설 : 출퇴근을 하는 30~50대가 가장 trip duration이 높을 것이다

#연령대 column
age = []
for year in df['birth_year']:
    if year > 1999:
        age.append(10)
    elif year <= 1999 and year > 1989:
        age.append(20)
    elif year <= 1989 and year > 1979:
        age.append(30)
    elif year <= 1979 and year > 1969:
        age.append(40)
    elif year <= 1969 and year > 1959:
        age.append(50)
    elif year <= 1959 and year > 1949:
        age.append(60)
    elif year <= 1949 and year > 1939:
        age.append(70)
    elif year <= 1939 and year > 1929:
        age.append(80)
    elif year <= 1929 and year > 1919:
        age.append(90)
    elif year <= 1919 and year > 1909:
        age.append(100)
    else:
        age.append(110)
df['age'] = age
plt.figure(figsize=(11,5))
box=sns.boxplot(x='age',y='trip_duration',data=df)
box.set_xlabel("연령대")
box.set_ylabel("주행 시간")
plt.ylim([0,4000])
plt.show()


```

**해석 보태기**
> 아웃라이어가 너무 많아 4000으로 y값을 제한하였음.
> 80세 이상의 duration time은 제외하고 생각했을 때,(아웃라이어라고 생각)
> 전 연령대 별 duration time의 차이는 크게 없었음. 
> 50대에서 타 연령대에 비해 살짝 더 높은 duration time 보임.

Bike_id, Trip_duration, + 한 가지 변수 더 추가해서 3개 이상 변수 시각화 방식으로 시각화 해주세요. 해석은 안 보태 주셔도 됩니다.
